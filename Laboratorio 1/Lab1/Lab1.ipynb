```python
#Imagen importada: 
img1_redbull = cv2.imread('rb.png')

#Correccion de BGR a RGB (matplotlib)
img1_redbull = cv2.cvtColor(img1_redbull,cv2.COLOR_BGR2RGB)

plt.imshow(img1_redbull)
```

```python
#Funcion creada
def imgColorSelection(img,color):
    
    #Dimensiones de la imagen
    alto = img.shape[0]
    ancho = img.shape[1]
    
    #Lienzo nuevo 
    lienzo = np.zeros((alto,ancho,3))
    
    #Bandera para seleccion nula de opciones
    band_null = 0
    
    #Recorrer la imagen 
    for i in range(0,alto):
        for j in range(0,ancho):
            pixel = img[i,j]
            
            #Canales 
            rojo = pixel[0]
            verde = pixel[1]
            azul = pixel[2]
            
            #Seleccion de colores 
            
            #Mostrar solo color azul
            if (color == 1):
                lienzo[i,j] = [azul,0,0]
            elif (color == 2):
                lienzo[i,j] = [0,verde,0]
            elif (color == 3):
                lienzo[i,j] = [0,0,rojo]
            elif (color == 10):
                lienzo[i,j] = [0,verde,rojo]
            elif (color == 20):
                lienzo[i,j] = [azul,verde,0]
            elif (color == 30):
                lienzo[i,j] = [azul,0,rojo]
            else: 
                band_null = 1 
    if band_null == 1: 
        print("No se selecciono ninguna opcion correcta")
    cv2.imwrite("problema1.jpg",lienzo)

#Test de la funcion
imgColorSelection(img1_redbull,1)
img_problema1 = cv2.imread('problema1.jpg')
img_problema1 = cv2.cvtColor(img_problema1,cv2.COLOR_BGR2RGB)
f, axarr = plt.subplots(1, 2, figsize=(20, 8))
axarr[0].imshow(img1_redbull, cmap='gray')
axarr[1].imshow(img_problema1, cmap='gray')
```

```python
#Funcion creada

def grayscaleToColor(img1,img2,img3):
    #Dimensiones de la imagen
    alto = img1.shape[0]
    ancho = img1.shape[1]

    #Lienzo nuevo 
    lienzo = np.zeros((alto,ancho,3))
    
    for i in range(0,alto):
        for j in range(0,ancho):
    
            #Canales 
            pixel1 = img1[i,j]
            pixel2 = img2[i,j]
            pixel3 = img3[i,j]
            
            #Colores 
            rojo = pixel1[0]
            verde = pixel2[1]
            azul = pixel3[2]
            
            lienzo[i,j] = [round(azul), round(verde), round(rojo)]
        
    cv2.imwrite("problema2.jpg",lienzo)

grayscaleToColor(perro_rojo,perro_verde,perro_azul)
img_problema2 = cv2.imread('problema2.jpg')
img_problema2 = cv2.cvtColor(img_problema2,cv2.COLOR_BGR2RGB)
                           
plt.imshow(img_problema2)
```

## Problema 3

```python
#Imagen importada: 
beagle = cv2.imread('beagle.jpg')

#Correccion de BGR a RGB (matplotlib)
beagle = cv2.cvtColor(beagle,cv2.COLOR_BGR2RGB)

plt.imshow(beagle)
```

```python
def colorToGrayscale(img):
    #Dimensiones de la imagen
    alto = img.shape[0]
    ancho = img.shape[1]

    #Lienzos nuevo 
    lienzo_r = np.zeros((alto,ancho,3))
    lienzo_g = np.zeros((alto,ancho,3))
    lienzo_b = np.zeros((alto,ancho,3))
    
    for i in range(0,alto):
        for j in range(0,ancho):
    
            #Canales 
            pixel = img[i,j]
        
            #Colores 
            rojo = pixel[0]
            verde = pixel[1]
            azul = pixel[2]
            
            #Creacion de lienzos en base al color
            lienzo_r[i,j] = [rojo,rojo,rojo]
            lienzo_g[i,j] = [verde,verde,verde]
            lienzo_b[i,j] = [azul,azul,azul]
            
    cv2.imwrite("problema3_r.jpg",lienzo_r)
    cv2.imwrite("problema3_g.jpg",lienzo_g)
    cv2.imwrite("problema3_b.jpg",lienzo_b)

colorToGrayscale(beagle)
img_problema3_r = cv2.imread('problema3_r.jpg')
img_problema3_g = cv2.imread('problema3_g.jpg')
img_problema3_b = cv2.imread('problema3_b.jpg')

img_problema3_r = cv2.cvtColor(img_problema3_r,cv2.COLOR_BGR2RGB)
img_problema3_g = cv2.cvtColor(img_problema3_g,cv2.COLOR_BGR2RGB)
img_problema3_b = cv2.cvtColor(img_problema3_b,cv2.COLOR_BGR2RGB)

f, axarr = plt.subplots(1, 4, figsize=(20, 8))
axarr[0].imshow(beagle, cmap='gray')
axarr[1].imshow(img_problema3_r, cmap='gray')
axarr[2].imshow(img_problema3_g, cmap='gray')
axarr[3].imshow(img_problema3_b, cmap='gray')
```

```python
#Funciones creadas para los histogramas 

def gray_hist(img):
    
    #Dimensiones
    alto = img.shape[0]
    ancho = img.shape[1]
    
    histograma = np.zeros([256],np.int32)
    
    #Llenar el histograma
    for i in range(0,alto):
        for j in range(0,ancho):
            histograma[img[i,j]]+=1
    plt.xlabel("Intensidad")
    plt.ylabel("Frecuencia")
    plt.plot(histograma,color='gray')

def color_hist(img):   
    
    #Dimensiones
    alto = img.shape[0]
    ancho = img.shape[1]
    
    urojo = 0
    uazul = 0
    uverde = 0

    histograma = np.zeros([256,3],np.int32)
    
    #LLenar el histograma
    for i in range(0,alto):
        for j in range(0,ancho):
            for c in range(0,3):
                pixel = img[i,j]
                if (c ==0):
                    uazul += pixel[0]
                elif (c==1):
                    uverde += pixel[1]
                elif (c==2):
                    urojo += pixel[2]
                    
                histograma[img[i,j,c],c]+=1
    urojo = urojo/(alto*ancho)
    uazul = uazul/(alto*ancho)
    uverde = uverde/(alto*ancho)

                
    plt.title("Histogramas")
    plt.xlabel("Intensidad")
    plt.ylabel("Frecuencia")
    plt.plot(histograma[:,0],'b')
    plt.plot(histograma[:,1],'g')
    plt.plot(histograma[:,2],'r')
    plt.axvline(x = urojo, color = 'red')
    plt.axvline(x = uazul, color = 'blue')
    plt.axvline(x = uverde, color = 'green')
    
color_hist(beagle)
gray_hist(beagle)
```

```python
#Imagen importada: 
pelota = cv2.imread('pelota.jpg')

#Correccion de BGR a RGB (matplotlib)
pelota = cv2.cvtColor(pelota,cv2.COLOR_BGR2RGB)

plt.imshow(pelota)
```

```python
#Funcion creada

def edgPonderado(img):
    #Dimensiones de la imagen
    alto = img.shape[0]
    ancho = img.shape[1]

    #Lienzo nuevo 
    lienzo = np.zeros((alto,ancho,3))
    
    for i in range(0,alto):
        for j in range(0,ancho):
    
            #Canales 
            pixel = img[i,j]
            
            #Colores 
            rojo = pixel[0]
            verde = pixel[1]
            azul = pixel[2]
            
            #Ponderado
            pond = (0.10*rojo)+(0.85*azul)+(0.05*verde)
            lienzo[i,j] = [round(pond), round(pond), round(pond)]
        
    cv2.imwrite("problema5.jpg",lienzo)

edgPonderado(pelota)
img_problema5 = cv2.imread('problema5.jpg')
img_problema5 = cv2.cvtColor(img_problema5,cv2.COLOR_BGR2RGB)
                           
f, axarr = plt.subplots(1, 2, figsize=(20, 8))
axarr[0].imshow(pelota, cmap='gray')
axarr[1].imshow(img_problema5, cmap='gray')
```
